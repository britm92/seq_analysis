```python
bed_df=pd.read_table("U00096.3.gtf",header=None)
bed_df[9]=bed_df[8].apply(lambda x: x.split(';')[4].split('=')[1])
bed_df=bed_df.loc[:,[3,4,9,6]]
bed_df.columns=['pos1','pos2','gene','strand']
```

```python
bed_df[bed_df['gene']=='yzgL']
```

```python
def get_subset(sample,bed_df,gene):
    gene_df=bed_df[bed_df['gene']==gene]
    start=int(gene_df["pos1"].values[0])
    end=int(gene_df["pos2"].values[0])
    
    
    
    path="2018-05-22_RNAseq/genome_cov_files/"
    fwd=pd.read_table(path+sample+"_fwd_perbase_cov.txt",header=None,index_col=0,usecols=[1,2],names=['loc','fwd_count'])
    rev=pd.read_table(path+sample+"_rev_perbase_cov.txt",header=None,index_col=0,usecols=[1,2],names=['loc','rev_count'])
    merge=fwd.join(rev)
    subset=merge.iloc[start:end,:]
    #subset=subset.replace(0,1)
    return subset
```

```python
gene=get_subset('bm03',bed_df,'yzgL')
```
